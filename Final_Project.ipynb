```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Import the dataset using Pandas from above mentioned url.

```python
Url = "https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv"
ds = pd.read_csv(Url)
ds
```

- Find no. of rows & columns in the dataset
- Data types of columns.

```python
print(f"Number of rows in the dataset is {ds.shape[0]}")
print(f"Number of columns in the dataset is {ds.shape[1]}")
```

## 3. Low Level Data Understanding :
- Find count of unique values in location column.
- Find which continent has maximum frequency using values counts.
- Find maximum & mean value in 'total_cases'.
- Find 25%,50% & 75% quartile value in 'total_deaths'.
- Find which continent has maximum 'human_development_index'.
- Find which continent has minimum 'gdp_per_capita'.

```python
print(f"Number of unique values in location column is {len(ds['location'].unique())}")
```

```python
dff = ds['continent'].value_counts()
print(f"The continent which has maximum frequency is {dff.index.to_list()[0]}")
```

```python
print(f"Maximum value in 'total_cases' is {ds['total_cases'].max()}")
```

```python
print(f"Mean value in 'total_cases' is {ds['total_cases'].mean()}")
```

## Find 25%,50% & 75% quartile value in 'total_deaths'.

```python
print(f"The 25% quartile value in 'total_deaths' is {ds['total_deaths'].quantile(q=0.25)}")
print(f"The 50% quartile value in 'total_deaths' is {ds['total_deaths'].quantile(q=0.5)}")
print(f"The 75% quartile value in 'total_deaths' is {ds['total_deaths'].quantile(q=0.75)}")
```

## Find which continent has maximum 'human_development_index'

```python
continent = ds.iloc[ds["human_development_index"].idxmax()]["continent"]
print(f"Continent which has maximum 'human_development_index' is {continent}")
```

## Find which continent has minimum 'gdp_per_capita'.

```python
continent2 = ds.iloc[ds["gdp_per_capita"].idxmin()]["continent"]
print(f"Continent which has minimum 'gdp_per_capita' is {continent2}")
```

## 4. Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.

```python
remove_clmns = ds.columns.difference(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']).to_list()
ds.drop(remove_clmns,axis=1,inplace=True)
ds
```

## 5. Data Cleaning

### a. Remove all duplicates observations

```python
ds.drop_duplicates(inplace=True)
```

### b. Find missing values in all columns

```python
ds.isnull().sum()
```

### c. Remove all observations where continent column value is missing

```python
ds.dropna(subset=['continent'],inplace=True)
```

### d. Fill all missing values with 0

```python
ds.fillna(0, inplace=True)
```

## 6. Date time format :

### a. Convert date column in datetime format using pandas.to_datetime

```python
pd.to_datetime(ds['date'])
```

### b. Create new column month after extracting month data from date column.


```python
ds['month'] = pd.to_datetime(ds['date']).dt.month
ds
```

## 7. Data Aggregation:

### a. Find max value in all columns using groupby function on 'continent' column

### b. Store the result in a new dataframe named 'df_groupby'.

```python
df_groupby = ds.groupby('continent').max().reset_index()
```

## 8. Feature Engineering :

### a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
ds['total_deaths_to_total_cases'] = ds['total_deaths']/ds['total_cases']
ds
```

## 9. Data Visualization :

### a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.distplot(df_groupby["gdp_per_capita"],bins=6)
```

### b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

```python
sns.scatterplot(x = 'total_cases', y = 'gdp_per_capita', data = df_groupby)
```

### c. Plot Pairplot on df_groupby dataset.

```python
sns.pairplot(data = df_groupby, diag_kind="kde")
```

### d. Plot a bar plot of 'continent' column with 'total_cases' .

```python
sns.catplot(x = 'continent', y ='total_cases', data=df_groupby, kind= 'bar', height= 8)
```
